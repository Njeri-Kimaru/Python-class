#### Sorting data
- we use a method called sort_values
- it has two parameters or arguments ;by and ascending

```python
# sort fare in ascending order
# only ascending never descending hence if descending then ascending = false
data = data.sort_values(by="Fare",ascending=False)
data.head()
```

```python
data = data.sort_values(by="Age",ascending=False)
data.head()
```

#### Feature engineering
- creating new columns that help improve your analysis and model


```python
#age_group = pd.cut
# when calling column start with table squqre brackets then name of column
# under pd.cut we have three arguments the column,bins and labels.
data["Age_group"] = pd.cut(data["Age"],bins=[0,12,19,59,100],labels=["child","teenager","adult","senior"])
data.head()
```

```python
data["Fare_rate"] = pd.cut(data["Fare"],bins=[0,200,600],labels=["cheap","expensive"])
data.head()
```

```python
# extracting titles from name
# APPLY ASSIGNMENT
# split according to index after the comma is the position one and before the fullstop its position 0
data["Title"]= data["Name"].apply(lambda name :name.split(",")[1].split(".")[0].strip())
data.head()
```
