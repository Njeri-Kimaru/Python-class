```python
# Use dictionary comprehension to combine the two lists and calculate the sample means drawn from the dictionary with the inputs being the sample size and the dataset i.e., our dictionary
# what is the difference between list comprehension and dictionary comprehension. Explain how you can combine two lists to come up with a dictionary comprehension
#List of people
people = [
    "Alice", "Bob", "Charlie", "Diana", "Ethan",
    "Fiona", "George", "Hannah", "Isaac", "Julia",
    "Kevin", "Laura", "Michael", "Nina", "Oscar"
]
heights = [
    165, 178, 172, 160, 185,
    170, 182, 158, 174, 169,
    180, 162, 176, 168, 181
]
```

```python
people_heights = {people[i]: heights[i] for i in range(len(people))}
people_heights
```

```python
import itertools
import numpy as np
from collections import Counter
import pandas as pd
```

```python
def group_mean(group_size,data):
    n = group_size
    mu = sum(data.values())/len(data)
    combinations = list(itertools.combinations(data.keys(), n))
    x_bar_grouplist = []
    
    for comb in combinations:
        people_heights = [data[p] for p in comb]
        group_mean = sum(people_heights)/n
        x_bar_grouplist.append(group_mean)
    return combinations, x_bar_grouplist
```

```python
def result_mu(x):
    w = sum(x.values())/len(x)
    return w
```

```python
people_heights_mu = result_mu(people_heights)
people_heights_mu
```

```python
# ", ".join --- removes tuples 
combinations , x_bar_grouplist = group_mean(3, people_heights)
people_heights = {", ".join(combinations[i]): x_bar_grouplist[i] for i in range(len(combinations))}
people_heights
```

```python
# separating the keys and the values
import pandas as pd
data = pd.DataFrame({
    "Group": list(people_heights.keys()),
    "Group_mean": list(people_heights.values())})
data.head()
```
