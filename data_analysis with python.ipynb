```python
# WRITE THE DATA IN EXCEL FILE TO SAMPLE.CSV
import csv

# Step 1: Open the source file (Excel.csv) for reading
with open('Excel_jumia 1.csv', mode='r', newline='', encoding='utf-8') as infile:
    reader = csv.reader(infile)

# Step 2: Open the destination file (sample.csv) for writing
    with open('sample.csv', mode='w', newline='', encoding='utf-8') as outfile:
        writer = csv.writer(outfile)

# Step 3: Copy each row from the source to the destination
        for row in reader:
            writer.writerow(row)
```
